## Converting Countinous variable to Numerical

```python
#Create a list for the columns to convert to numeric
columns_to_convert = ['price', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'normalized-losses']

# convert them using pd.to_numeric formula
motor[columns_to_convert] = motor[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

## Handling missing values in normalized losses, horsepower, price etc.

- with their respective means and convert them to data types
- Provide before and after summaries

```python
## Summary of the dataframe before handling missing values
motor.describe()
```

```python
for col in columns_to_convert:
    if col in motor.columns:
        mean_val = motor[col].mean()
        motor[col] = motor[col].fillna(mean_val)
        print(f"{col} column has a mean of {mean_val:.2f}")
```

```python
# Fill missing values in 'num-of-doors' with the mode
motor['num-of-doors'] = motor['num-of-doors'].fillna(motor['num-of-doors'].mode()[0])

print(f"'num-of-doors' column has been converted to mode: {motor['num-of-doors'].mode()[0]}")
```
