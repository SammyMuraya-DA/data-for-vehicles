## Convert categorical data (if any) into numerical format using appropriate encoding techniques

Converting categorical variables into numerical format is essential for most machine learning algorithms, as they work better with numerical data.

## Choosing encoding technique
Depending on the column type (nominal or ordinal), we can decide which encoding technique to use:
   - One-Hot Encoding for nominal data (no inherent order).
   -  Label Encoding for ordinal data (ordered categories).
   - Binary Encoding for high-cardinality categorical variables.
   - Target Encoding if categories strongly relate to the target variable (like price).

From the data, there are different groups of categorical data. The num-of-doors and
num-of-cylinders have number categories in string. The other categories has non numeric norminal

```python
motor_data_encoded = motor.copy()
motor_data_encoded['n_doors'] = motor_data_encoded['num-of-doors'].map({'two':2, 'four':4}).astype('int')
motor_data_encoded['n_cylinders'] = motor_data_encoded['num-of-cylinders'].map({
    'two':2, 'three':3, 'four':4, 'five':5,'six':6, 'eight':8, 'twelve':12}).astype('int')
motor_data_encoded.head()
```

```python
# Droping multiple columns of the same values
# Transformed data
motor_transformed = motor_data_encoded.drop(columns=['num-of-doors', 'num-of-cylinders', 'z_score_price', 'price'])
motor_transformed.head()
```

## Correlation Matrix

Create a correlation matrix and identify the strongest and weakest correlations

```python
# Select numeric columns from the transformed DataFrame
numeric_data = motor_transformed.select_dtypes(include=['float', 'int'])

# Calculate the correlation matrix, round to 4 decimal places, and capitalize column names
corr_matrix = numeric_data.corr().round(4).rename(columns=str.title, index=str.title)

# Display the correlation matrix
corr_matrix
```

```python
# Unstack and sort correlation matrix
unstacked_corr = corr_matrix.unstack().sort_values()

# Find the strongest positive and negative correlations, excluding self-correlations (value of 1)
strong_pos_value = unstacked_corr[unstacked_corr != 1].max()
strong_pos_comb = unstacked_corr[unstacked_corr != 1].idxmax()  
strong_neg_comb = unstacked_corr[unstacked_corr != 1].idxmin()
strong_neg_val = unstacked_corr[unstacked_corr != 1].min()

# Display results
print("Strong Correlations\n" + "_" * 50)
if strong_pos_value > 0.8 and strong_neg_val < -0.8:
    print(f"| Pos {strong_pos_comb}: {strong_pos_value:.2f}")
    print(f"| Neg {strong_neg_comb}: {strong_neg_val:.2f}")
elif strong_pos_value > 0.8:
    print(f"| Pos {strong_pos_comb}: {strong_pos_value:.2f}")
    print("| There is no strong negative correlation")
elif strong_neg_val < -0.8:
    print("| There is no strong positive correlation")
    print(f"| Neg {strong_neg_comb}: {strong_neg_val:.2f}")
else:
    print("| There is no strong correlation")
```

```python
# Separate positive and negative correlations
unstacked_corr_neg = unstacked_corr[unstacked_corr < 0]
unstacked_corr_pos = unstacked_corr[unstacked_corr > 0]

# Find the weakest positive and negative correlations
weak_pos_comb = unstacked_corr_pos.idxmin()
weak_pos_value = unstacked_corr_pos.min()
weak_neg_comb = unstacked_corr_neg.idxmax()
weak_neg_val = unstacked_corr_neg.max()

# Display results
print("Weak Correlations\n" + "_" * 80)
print(f"Positive -> {weak_pos_comb[0]} and {weak_pos_comb[1]}: {weak_pos_value:.2f}")
print(f"Negative -> {weak_neg_comb[0]} and {weak_neg_comb[1]}: {weak_neg_val:.2f}")
```

## Insights from Correlation Results:

# Strong Correlations:

   - Positive Correlation (Highway-Mpg and City-Mpg: 0.97)
        There is a very strong positive correlation between highway and city mileage. This indicates that cars with high mileage on highways also tend to have high mileage in city driving, suggesting fuel efficiency is consistent across different driving conditions.

   - Negative Correlation (City-Mpg and Horsepower: -0.80)
        There is a strong negative correlation between city mileage and horsepower. This means that cars with higher horsepower typically have lower city mileage, indicating that powerful engines consume more fuel during city driving.

# Weak Correlations:

   - Positive Correlation (Bore and Compression-Ratio: 0.01)
        The negligible positive correlation suggests that bore size and compression ratio have almost no linear relationship, implying changes in bore size do not significantly affect the compression ratio.

   - Negative Correlation (Stroke and Number of Doors: -0.01)
        This near-zero negative correlation indicates that the engine stroke length and the number of doors are virtually unrelated, showing no significant connection between these features.
